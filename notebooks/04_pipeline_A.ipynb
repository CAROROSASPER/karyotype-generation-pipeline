```python
from google.colab import drive
drive.mount('/content/drive')

import os, json, math, shutil
from pathlib import Path
import numpy as np, cv2, pandas as pd
from tqdm import tqdm
import torch
from ultralytics import YOLO

# ── CONFIG ────────────────────────────────────────────────────────────
ROOT     = "/content/drive/MyDrive/54816"
OUT_DIR  = f"{ROOT}/e2e_run_pipeline_A"
os.makedirs(OUT_DIR, exist_ok=True)

OVERLAP_W     = "/content/drive/MyDrive/Modelos/overlap_modelo_entrenado2.pt"
YOLO24_W      = f"{ROOT}/weight/best-24_chromosomes.pt"
AE_KERAS      = f"{ROOT}/weights_autoencoder/autoencoder_model.keras"

TARGET_DIR    = f"{ROOT}/24_chromosomes_object/JEPG"

OVERLAP_LABEL = "overlapped"
OVERLAP_CONF  = 0.25
YOLO_CONF     = 0.25
YOLO_IOU      = 0.70
AE_IMG_SIZE   = (128,128)

KARYO_ORDER = [
    "A1","A2","A3","B4","B5",
    "C6","C7","C8","C9","C10","C11","C12",
    "D13","D14","D15","E16","E17","E18",
    "F19","F20","G21","G22","X","Y"
]

DEVICE = 0 if torch.cuda.is_available() else "cpu"

def exists(p): print("✅" if Path(p).exists() else "❌", p)
for p in [OVERLAP_W, YOLO24_W, TARGET_DIR]: exists(p)
print("AE:", "✅" if Path(AE_KERAS).exists() else "⚠️ optional", AE_KERAS)
```

```python
# ── Utilities ─────────────────────────────────────────────────────────
IMG_EXTS = {".jpg",".jpeg",".png",".bmp",".tif",".tiff"}

def list_images(d):
    files=[]
    for e in IMG_EXTS: files+=list(Path(d).rglob(f"*{e}")); files+=list(Path(d).rglob(f"*{e.upper()}"))
    return sorted(set(files))

def imread_rgb(p):
    bgr=cv2.imread(str(p),cv2.IMREAD_COLOR)
    return cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB) if bgr is not None else None

def crop_bbox(img,x1,y1,x2,y2,pad=0):
    h,w=img.shape[:2]
    x1=max(0,int(x1)-pad); y1=max(0,int(y1)-pad)
    x2=min(w,int(x2)+pad); y2=min(h,int(y2)+pad)
    return img[y1:y2,x1:x2].copy() if x2>x1 and y2>y1 else None

def numerical_alert(counts):
    exp={k:2 for k in KARYO_ORDER if k not in ("X","Y")}
    exp["X"]=1 if counts.get("Y",0)>0 else 2
    exp["Y"]=1 if counts.get("Y",0)>0 else 0
    return {k:{"obs":int(counts.get(k,0)),"exp":int(e)}
            for k,e in exp.items() if int(counts.get(k,0))!=int(e)}

def build_karyotype(crops_by_label, out_path, cell=(220,220), cols=4, pad=10):
    cw,ch=cell; order=KARYO_ORDER; n=len(order)
    rows=math.ceil(n/cols)
    canvas=np.ones((rows*ch+(rows+1)*pad, cols*cw+(cols+1)*pad,3),np.uint8)*255
    for idx,lbl in enumerate(order):
        r,c=idx//cols,idx%cols
        y0=pad+r*(ch+pad); x0=pad+c*(cw+pad)
        imgs=crops_by_label.get(lbl,[])
        cell_img=np.ones((ch,cw,3),np.uint8)*255
        if imgs:
            half=(ch-10)//2
            def fit(img,th):
                h,w=img.shape[:2]; s=min(th/h,cw/w)
                return cv2.resize(img,(max(1,int(w*s)),max(1,int(h*s))),cv2.INTER_AREA)
            if len(imgs)==1:
                t=fit(imgs[0],ch); oy=(ch-t.shape[0])//2; ox=(cw-t.shape[1])//2
                cell_img[oy:oy+t.shape[0],ox:ox+t.shape[1]]=t
            else:
                for i,(img,y_off) in enumerate(zip(imgs[:2],[0,half+10])):
                    t=fit(img,half); ox=(cw-t.shape[1])//2
                    cell_img[y_off:y_off+t.shape[0],ox:ox+t.shape[1]]=t
        canvas[y0:y0+ch,x0:x0+cw]=cell_img
    cv2.imwrite(str(out_path),cv2.cvtColor(canvas,cv2.COLOR_RGB2BGR))
    return out_path
```

```python
# ── Stage 1: Overlap gate ─────────────────────────────────────────────
overlap_model = YOLO(OVERLAP_W)
names = overlap_model.names
id2n  = names if isinstance(names,dict) else {i:n for i,n in enumerate(names)}
n2id  = {str(v).strip().lower():int(k) for k,v in id2n.items()}
ov_id = n2id[OVERLAP_LABEL.lower()]

all_imgs = list_images(TARGET_DIR)
kept, discarded = [], []
for p in tqdm(all_imgs, desc="Overlap gate"):
    r=overlap_model(str(p),verbose=False,device=DEVICE,conf=OVERLAP_CONF)[0]
    ids=[int(c) for c in r.boxes.cls] if r.boxes and len(r.boxes) else []
    (discarded if ov_id in ids else kept).append(p)

print(f"Kept {len(kept)}/{len(all_imgs)} images after overlap gate.")
pd.DataFrame({"image":[p.name for p in all_imgs],
              "kept":[p in set(kept) for p in all_imgs]}
             ).to_csv(os.path.join(OUT_DIR,"overlap_gate.csv"),index=False)
```

```python
# ── Stages 2-4: YOLO 24-class → count validation → karyotype ─────────
yolo24 = YOLO(YOLO24_W)
karyo_dir = Path(OUT_DIR)/"karyotypes_A"; karyo_dir.mkdir(exist_ok=True)

results=[]
for img_path in tqdm(kept, desc="Pipeline A"):
    img=imread_rgb(img_path)
    if img is None: continue
    r=yolo24(str(img_path),verbose=False,device=DEVICE,conf=YOLO_CONF,iou=YOLO_IOU)[0]
    id2lbl=r.names if isinstance(r.names,dict) else {i:n for i,n in enumerate(r.names)}
    counts={}; crops_by_label={}
    if r.boxes and len(r.boxes):
        for box in r.boxes:
            lbl=id2lbl[int(box.cls[0])]
            x1,y1,x2,y2=box.xyxy[0].tolist()
            crop=crop_bbox(img,x1,y1,x2,y2)
            if crop is None: continue
            counts[lbl]=counts.get(lbl,0)+1
            crops_by_label.setdefault(lbl,[]).append(crop)

    alerts=numerical_alert(counts)
    karyo_path=karyo_dir/f"{img_path.stem}_A.png"
    build_karyotype(crops_by_label, karyo_path)

    results.append({"image":img_path.name,"counts":counts,
                    "alerts":alerts,"karyotype":str(karyo_path)})

print(f"Processed {len(results)} images. Karyotypes saved to {karyo_dir}")
```

```python
# ── Stage 5 (optional): Autoencoder anomaly scoring ───────────────────
if Path(AE_KERAS).exists():
    from tensorflow.keras.models import load_model
    ae=load_model(AE_KERAS); print("✅ AE loaded")

    def ae_score(crop):
        x=cv2.resize(crop,AE_IMG_SIZE).astype(np.float32)/255.0
        xb=x[None]; xhat=ae.predict(xb,verbose=0)[0]
        return float(np.mean((x-xhat)**2))

    # Fit threshold from all crops
    all_crops=[]
    for entry in results:
        img=imread_rgb(Path(TARGET_DIR)/entry["image"])
        if img is None: continue
        r=yolo24(str(Path(TARGET_DIR)/entry["image"]),verbose=False,device=DEVICE,
                 conf=YOLO_CONF,iou=YOLO_IOU)[0]
        id2lbl=r.names if isinstance(r.names,dict) else {i:n for i,n in enumerate(r.names)}
        if r.boxes and len(r.boxes):
            for box in r.boxes:
                x1,y1,x2,y2=box.xyxy[0].tolist()
                c=crop_bbox(img,x1,y1,x2,y2)
                if c is not None: all_crops.append(c)

    threshold=float(np.percentile([ae_score(c) for c in all_crops],95)) if all_crops else 0.02
    print(f"AE threshold (p95): {threshold:.6f}")

    for entry in results:
        img=imread_rgb(Path(TARGET_DIR)/entry["image"])
        if img is None: continue
        r=yolo24(str(Path(TARGET_DIR)/entry["image"]),verbose=False,device=DEVICE,
                 conf=YOLO_CONF,iou=YOLO_IOU)[0]
        id2lbl=r.names if isinstance(r.names,dict) else {i:n for i,n in enumerate(r.names)}
        ae_flags={}
        if r.boxes and len(r.boxes):
            for box in r.boxes:
                lbl=id2lbl[int(box.cls[0])]
                x1,y1,x2,y2=box.xyxy[0].tolist()
                c=crop_bbox(img,x1,y1,x2,y2)
                if c is None: continue
                err=ae_score(c)
                if err>ae_flags.get(lbl,{}).get("max_err",0):
                    ae_flags[lbl]={"max_err":err,"anomaly":err>threshold}
        entry["ae_scores"]=ae_flags
    print("✅ AE scoring complete")
else:
    print("⚠️  AE model not found — skipping anomaly scoring.")
```

```python
# ── Save consolidated results ─────────────────────────────────────────
import json
out_path=os.path.join(OUT_DIR,"pipeline_a_results.json")
with open(out_path,"w") as f: json.dump(results,f,indent=2,ensure_ascii=False,default=str)
print("Results saved to:", out_path)

alerts_count=sum(1 for r in results if r["alerts"])
print(f"Images with numerical alerts: {alerts_count}/{len(results)}")
```
